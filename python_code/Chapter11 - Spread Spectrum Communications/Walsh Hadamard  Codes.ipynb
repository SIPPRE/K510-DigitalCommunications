```python
import numpy as np
from scipy.linalg import hadamard
import matplotlib.pyplot as plt

def generate_walsh_codes(n=8):
    """Generate Walsh-Hadamard codes"""
    H = hadamard(n)
    return H

def visualize_walsh_codes():
    # Generate 8x8 Walsh codes
    walsh_codes = generate_walsh_codes(16)

    # Visualize codes
    plt.figure(figsize=(15, 10))

    # Matrix visualization
    plt.subplot(121)
    plt.imshow(walsh_codes, cmap='bwr')
    plt.colorbar(label='Amplitude')
    plt.title('Walsh-Hadamard Codes (8x8 Matrix)')
    plt.xlabel('Code Index')
    plt.ylabel('Chip Index')

    # Individual code sequences
    plt.subplot(122)
    for i in range(16):
        plt.plot(walsh_codes[:, i] + 2*i, label=f'Code {i+1}')
    plt.title('Individual Walsh Code Sequences')
    plt.xlabel('Chip Index')
    plt.ylabel('Code Sequence (offset for clarity)')
    plt.grid(True)
    plt.legend()

    plt.tight_layout()
    plt.show()

    # Print the codes
    print("\nWalsh-Hadamard Codes (8x8):")
    print(walsh_codes)

    # Verify orthogonality
    correlation = walsh_codes.T @ walsh_codes / 8
    print("\nCross-correlation Matrix:")
    print(correlation)

visualize_walsh_codes()
```
